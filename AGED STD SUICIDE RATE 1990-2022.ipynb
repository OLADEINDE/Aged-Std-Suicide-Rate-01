# What was the average DeathRatePer100k in each RegionaName (Continents) per year ?

```python
df['DeathRatePer100K'].mean()
```

# What was the average DeathRatePer100K across all continents (RegionName) per year ?

```python
df.groupby('Year')['DeathRatePer100K'].mean()
```

# What was the average DeathRatePer100K for Each distinct continents (RegionName) per year ?


```python
df.pivot_table(index='Year', columns='RegionName',values='DeathRatePer100K').dropna()
```

df.plot()

```python
df.pivot_table(index='Year', columns=('Sex','RegionName'),values='DeathRatePer100K').dropna()
```

```python
df.pivot_table(index='Year', columns='RegionName',aggfunc='sum',values='DeathRatePer100K').dropna()
```

```python
df.aggregate('Population').dropna()
```

# What is the TOTAL suicide count in each in each continents (RegionName)

```python
df2= df.groupby('RegionName')['SuicideCount'].mean()
df2
```

```python
df2.plot(kind='line',title='Suicide Count Across World Continents', ylabel='SuicideCount',xlabel='RegionName')
```
